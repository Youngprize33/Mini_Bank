## 목표금액 기준 단리 및 복리 적금계산기 만들기

### 목표금액 기준 단리 적금계산기

* 입력 : 목표금액, 목표기간, 연금리

```python
목표금액 = 100000000
목표기간 = 120
연금리 = 4
연금리 = 연금리 / 100
월금리 = 연금리 / 12
```

```python
월금리
```

```python
Image('data/목표금액기준적금단리계산공식.JPG')
```

```python
납입원금 = 목표금액 / (1 + (목표기간 + 1)/2 * 월금리 * (1 - 일반세율))
납입원금
```

```python
세전이자 = 납입원금 * (목표기간 + 1)/2 * 월금리
세전이자
```

```python
세금 = 세전이자 * 일반세율
세금
```

```python
월납입금 = 납입원금 / 목표기간
월납입금
```

```python
월납입금 = round(월납입금)
월납입금
```

```python
월납입금 = format(월납입금, ',')
월납입금
```

## 목표금액 기준 복리 적금계산기

```python
Image('data/목표금액기준적금복리계산공식.JPG')
```

* 월납입금, 납입원금 계산

```python
월납입액 = 목표금액/((((1+월금리)**목표기간-일반세율*(1+월금리)**목표기간)-1+일반세율)/(월금리/(1+월금리))+목표기간*일반세율)
월납입액
```

```python
납입원금 = 월납입액 * 목표기간
납입원금
```

```python
월납입액 = round(월납입액)
월납입액 = format(월납입액, ',')
월납입액
```

```python
납입원금 = round(납입원금)
납입원금 = format(납입원금, ',')
납입원금
```
