# Input 함수 <br>
* input 함수는 표준입력 환경 제공
* 입력된 값을 변수에 저장하여 사용
* 변수명 = input('입력레이블')

# Round 함수 <br>
* 변수에 저장된 값에 대해 소수점 이하 자리수 반올림
* 소수점 이하 값 저장 변수와 반올림 자리수
* 변수 = round(변수, 자리수)

# 단리예금계산 <br>
* 세전이자 = 예치금액 * 월금리 * 예치기간(월)
* 세금 = 세전이자 * 세금(일반, 우대, 비과세)
* 세후이자 = 세전이자 - 금
* 만기지급금 = 예치금액 + 세후이자

* 입력 : 예치금액, 예치기간, 연금리

```python
예치금액 = input('예치금액 : ')
```

```python
예치금액 # 문자열로 저장됨
```

```python
예치기간 = input('예치기간 : ')
```

```python
예치기간
```

```python
연금리 = input('연금리(%) : ')
```

```python
연금리
```

* 세율 변수 생성 및 저장

```python
일반세율 = 0.154
우대세율 = 0.095
비과세세율 = 0
```

* 일반세율 계산

```python
세전이자 = 예치금액 * 월금리 * 예치기간
세전이자
```

```python
세금 = 세전이자 * 일반세율
세금
```

```python
만기지급액 = 예치금액 + 세후이자
만기지급액
```

* round 함수를 이용한 반올림

```python
round(만기지급액)
```

* 세후이자 = 세전이자 - 세금

```python
세금 = 세전이자 * 일반세율
세금
```
